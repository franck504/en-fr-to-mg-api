# Backend EN/FR -> MG sur Google Colab

Ce notebook recupere le projet depuis GitHub, lance le backend FastAPI directement dans Colab et expose l'API via ngrok.

Avant d'executer les cellules:
- creez un depot GitHub pour ce projet
- remplacez l'URL du depot dans la cellule ci-dessous
- creez un token gratuit sur ngrok et remplacez la valeur plus bas
- creez une cle API Gemini gratuite dans Google AI Studio si vous voulez tester le provider Gemini


```python
REPO_URL = "https://github.com/franck504/en-fr-to-mg-api.git"
BRANCH = "main"
PROJECT_DIR = "/content/en_fr_to_MG"
```

```python
import os
import subprocess

if os.path.exists(PROJECT_DIR):
    subprocess.run(["git", "-C", PROJECT_DIR, "fetch", "origin"], check=True)
    subprocess.run(["git", "-C", PROJECT_DIR, "checkout", BRANCH], check=True)
    subprocess.run(["git", "-C", PROJECT_DIR, "pull", "origin", BRANCH], check=True)
else:
    subprocess.run(["git", "clone", "--branch", BRANCH, REPO_URL, PROJECT_DIR], check=True)

os.chdir(PROJECT_DIR)
print(f"Project synced in {PROJECT_DIR}")
```
